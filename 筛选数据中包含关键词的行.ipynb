```python
import pandas as pd

# 读取数据文档，假设数据文档名为data.csv，可以根据实际文件名修改
data = pd.read_csv('标注和适用车型的遮阳板(套装)(改进).csv')

# 定义关键词列表
keywords = [
    r'\(3-20\)pcs', r'\(3-20\) pcs', r'\(3-20\) pieces',
    r'\(3-20\)pieces', r'\(3-20\) pcs', r'\(3-20\)pcs',
    r'\(3-20\) pack', r'\(3-20\)pack', 'set'
]

# 将关键词列表合并为正则表达式，不区分大小写
pattern = '|'.join(keywords)

# 使用str.contains方法筛选包含关键词的行
filtered_data = data[data['商品标题'].str.contains(pattern, case=False, regex=True)]

# 输出筛选结果
print(filtered_data)

# 输出筛选结果到新的CSV文件
filtered_data.to_csv('标注和适用车型的遮阳板(已套装)(改进).csv', index=False)
```
