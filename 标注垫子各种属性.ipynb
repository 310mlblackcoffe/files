```python




import pandas as pd
import re

# 读取excel文件
file_path = '12.16-1.16垫子全量（标注车型）（剔除通用）.xlsx'  # 替换成你的文件路径
df = pd.read_excel(file_path)

# 定义一个函数来标注座椅数量
# 暂被注销

"""

def label_seat_number(title):
    labels = set()  # 使用集合来确保座椅数量不重复
    lower_title = title.lower()  # 将标题转换为小写
    
    if any((keyword in lower_title) for keyword in ["2nd row bench seat", "7 seat", "7 seater", "7-seater", "7-seat", "7-seat", "7seat", "7 passenger", "7 seating", "7-passenger", "seven seater"]):
        labels.add('7座')
    if any((keyword in lower_title) for keyword in ["8 seat", "8 seater", "8-seater", "8-seat", "8-seat", "8seat", "8 passenger", "8 seating", "8-passenger", "eight seater"]):
        labels.add('8座')
    if any((keyword in lower_title) for keyword in ["2nd row bucket seat", "6 seat", "6 seater", "6-seater", "6-seat", "6-seat", "6seat", "6 passenger", "6 seating", "6-passenger", "six seater"]):
        labels.add('6座')
    
    return '|'.join(labels) if labels else ''

"""

# 定义一个函数来标注座椅形状
# 暂被注销

"""

def label_seat_style(title):
    lower_title = title.lower()  # 将标题转换为小写
    labels = []
    if any((keyword in lower_title) for keyword in ["1st row bench seat", "front row bench seat", "1st row captain seat"]):
        labels.append('第一排3座')
    elif any((keyword in lower_title) for keyword in ["1st row bucket or bench seat"]):
        labels.append('第一排2-3座')
    elif any((keyword in lower_title) for keyword in ["1st / 2nd row bench seat"]):
        labels.append('一二排都是长椅')
    elif any((keyword in lower_title) for keyword in ["2nd row bench seat"]):
        labels.append('第二排3座')
    elif any((keyword in lower_title) for keyword in ["2nd row bucket seat", "2nd row captain seat"]):
        labels.append('第二排2座')
    elif any((keyword in lower_title) for keyword in ["bench seat"]):
        labels.append('长椅')
    elif any((keyword in lower_title) for keyword in ["bucket seat"]):
        labels.append('斗式座椅')
    elif any((keyword in lower_title) for keyword in ["captain seat"]):
        labels.append('船长椅')
    
    return '|'.join(labels) if labels else ''
    
"""
    


# 定义一个函数来标注适用配置
# 暂被注销

"""
def label_configuration(title):
    labels = set()  # 使用集合来确保座椅数量不重复
    lower_title = title.lower()  # 将标题转换为小写

    if any((keyword in lower_title) for keyword in ["plug-in hybrid"]):
        labels.add('插电式混动')
    elif any((keyword in lower_title) for keyword in ["hybrid"]):
        labels.add('混动')
    elif any((keyword in lower_title) for keyword in ["non-hybrid", "no hybrid", "non-hybrid", "non hybrid", "not hybrid", "not fit hybrid", "not for hybrid"]):
        labels.add('不适用混动')
    elif any((keyword in lower_title) for keyword in ["not plug in", "not for plug-in", "not for plug in", "not for phev", "no phev", "not fit plug-in"]):
        labels.add('不适用插电式混动')
    elif any((keyword in lower_title) for keyword in ["for awd"]):
        labels.add('四驱')
    elif any((keyword in lower_title) for keyword in ["non awd", "not fit for fwd", "not fit fwd", "no awd", "not for awd"]):
        labels.add('不适配四驱')

    return '|'.join(labels) if labels else ''

"""


# 定义一个函数来标注是否适配低音炮
# 暂被注销

"""
def label_subwoofer(title):
    pattern_subwoofer_1 = r'\b(?:with subwoofer|w/ subwoofer|with factory subwoofer|w/ factory subwoofer|w/factory subwoofer|with sub woofer|with speaker)\b'
    pattern_subwoofer_2 = r'\b(?:subwoofer|sub woofer|speaker)\b'
    pattern_subwoofer_3 = r'\b(?:w/o|no|not|without)\b'
    
    if re.search(pattern_subwoofer_1, title, re.IGNORECASE):
        return '适配低音炮'
    if re.search(pattern_subwoofer_2, title, re.IGNORECASE) and re.search(pattern_subwoofer_3, title, re.IGNORECASE):
        return '不适配低音炮'
    
    return ''
    
"""




# 定义一个函数来标注商品名称是否为脚垫
def label_floor_mats_name(title):
    pattern_floor_mats_1 = r'\b(?:floor|floorliner|floorliners|floormats|front liners|3rd row|2nd row|2nd rows|1st row|front row|3rd rows|3 rows|row liner|row liners|slush|smartliner|husky|max liner|all season|all weather|all-weather)\b'
    pattern_floor_mats_2 = r'\b(?:floorliner|floorliners|liner|liners|mats|mat|smartliner)\b'
    
    if re.search(pattern_floor_mats_1, title, re.IGNORECASE) and re.search(pattern_floor_mats_2, title, re.IGNORECASE):
        return '脚垫'
    
    return ''



# 定义一个函数来标注货箱垫上下层
def label_cargo_location(title):
    pattern_cargo_location_mats_1 = r'\b(?:cargo|trunk)\b'
    pattern_cargo_location_mats_2 = r'\b(?:liner|liners|mats|mat)\b'
    pattern_cargo_location_mats_3 = r'\b(?:lower)\b'
    pattern_cargo_location_mats_4 = r'\b(?:upper)\b'


    if re.search(pattern_cargo_location_mats_1, title, re.IGNORECASE) and re.search(pattern_cargo_location_mats_2, title, re.IGNORECASE) and re.search(pattern_cargo_location_mats_3, title, re.IGNORECASE) and re.search(pattern_cargo_location_mats_4, title, re.IGNORECASE):
        return '货厢上下层垫'
    if re.search(pattern_cargo_location_mats_1, title, re.IGNORECASE) and re.search(pattern_cargo_location_mats_2, title, re.IGNORECASE) and re.search(pattern_cargo_location_mats_3, title, re.IGNORECASE):
        return '货厢下层垫'
    if re.search(pattern_cargo_location_mats_1, title, re.IGNORECASE) and re.search(pattern_cargo_location_mats_2, title, re.IGNORECASE) and re.search(pattern_cargo_location_mats_4, title, re.IGNORECASE):
        return '货厢上层垫'
    
    return ''




# 定义一个函数来标注商品名称是否为货厢垫
def label_cargo_liner_name(title):
    pattern_cargo_liner_1 = r'\b(?:front cargo|front trunk)\b'
    pattern_cargo_liner_2 = r'\b(?:cargo|trunk)\b'
    pattern_cargo_liner_3 = r'\b(?:liner|liners|mats|mat|tray|protector)\b'

    
    if re.search(pattern_cargo_liner_1, title, re.IGNORECASE) and re.search(pattern_cargo_liner_3, title, re.IGNORECASE) :
        return '前备箱垫'
    if re.search(pattern_cargo_liner_2, title, re.IGNORECASE) and re.search(pattern_cargo_liner_3, title, re.IGNORECASE):
        return '后备箱垫'

    return ''


# 定义一个函数来标注商品名称是否为椅背垫
def label_backrest_mats_name(title):
    pattern_backrest_mats_1 = r'\b(?:backrest|seat back|seats back)\b'
    pattern_backrest_mats_2 = r'\b(?:liner|liners|mats|mat|cover|covers|protector)\b'
    
    if re.search(pattern_backrest_mats_1, title, re.IGNORECASE) and re.search(pattern_backrest_mats_2, title, re.IGNORECASE):
        return '椅背垫'
    
    return ''



# 定义一个函数来标注商品名称是否为门槛条
def label_door_sill_name(title):
    pattern_door_sill_1 = r'\b(?:door sill|door edge)\b'
    pattern_door_sill_2 = r'\b(?:guards|guard|protector|protectors)\b'
    
    if re.search(pattern_door_sill_1, title, re.IGNORECASE) and re.search(pattern_door_sill_2, title, re.IGNORECASE):
        return '门槛条'
    
    return ''





# 定义一个函数来标注商品名称是否为后斗垫
def label_bed_mat_name(title):
    pattern_bed_mat_1 = r'\b(?:bed mat|bed liner|bed mats|bedmat|bedrug)\b'
    
    if re.search(pattern_bed_mat_1, title, re.IGNORECASE):
        return '后斗垫'
    
    return ''




# 定义一个函数来标注商品名称是否为尾门台阶垫
def label_tailgate_cap_name(title):
    pattern_tailgate_cap_1 = r'\b(?:tailgate)\b'
    pattern_tailgate_cap_2 = r'\b(?:mold|mould|molding|moulding)\b'
    pattern_tailgate_cap_3 = r'\b(?:cap|caps)\b'
    pattern_tailgate_cap_exclude = r'\b(?:clips)\b'


    if re.search(pattern_tailgate_cap_1, title, re.IGNORECASE) and re.search(pattern_tailgate_cap_2, title, re.IGNORECASE) and re.search(pattern_tailgate_cap_3, title, re.IGNORECASE) and not re.search(pattern_tailgate_cap_exclude, title, re.IGNORECASE):
        return '尾门台阶垫'
    
    return ''





# 定义一个函数来标注商品名称是否为尾门垫
def label_tailgate_mat_name(title):
    pattern_tailgate_mat_1 = r'\b(?:tailgate mat|tailgate pad|tailgate liner)\b'
    
    if re.search(pattern_tailgate_mat_1, title, re.IGNORECASE):
        return '尾门垫'
    
    return ''



# 在dataframe中添加新的列
# df['座椅数量'] = df['商品标题'].apply(label_seat_number)
# df['座椅形状'] = df['商品标题'].apply(label_seat_style)
# df['低音炮'] = df['商品标题'].apply(label_subwoofer)
# df['适用配置'] = df['商品标题'].apply(label_configuration)
df['脚垫'] = df['商品标题'].apply(label_floor_mats_name)
df['货厢垫上下层'] = df['商品标题'].apply(label_cargo_location)
df['后备箱垫'] = df['商品标题'].apply(label_cargo_liner_name)
df['椅背垫'] = df['商品标题'].apply(label_backrest_mats_name)
df['门槛条'] = df['商品标题'].apply(label_door_sill_name)
df['后斗垫'] = df['商品标题'].apply(label_bed_mat_name)
df['尾门台阶垫'] = df['商品标题'].apply(label_tailgate_cap_name)
df['尾门垫'] = df['商品标题'].apply(label_tailgate_mat_name)




# 导出带有新列的excel文件
output_file_path = '12.16-1.16垫子全量（标注车型）（剔除通用）（标注属性-2）.xlsx'  # 替换成你想要的输出文件路径
df.to_excel(output_file_path, index=False)

print("处理完成，已导出到", output_file_path)
```
